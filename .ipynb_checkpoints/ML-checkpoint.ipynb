### Feature Engineering 
*Feature engineering refers to the process of using domain knowledge to select and transform the most relevant variables from raw data when creating a predictive model using machine learning or statistical modeling*

```python
df.dtypes
```

*We'll select the numerical columns in-order to predict regression based model*

```python
numeric_lst=['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numerical_cols = list(df.select_dtypes(include=numeric_lst).columns)
```

```python
numerical_cols
```

```python
data = df[numerical_cols]
```

```python
data.dtypes
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data.drop(labels=['Life expectancy '], axis=1),data['Life expectancy '],test_size=0.3,random_state=0)
```

```python
X_train.isnull().sum()
```

```python
from sklearn.feature_selection import mutual_info_regression
# determine the mutual information
mutual_info = mutual_info_regression(X_train, y_train)
mutual_info
```

```python
mutual_info = pd.Series(mutual_info)
mutual_info.index = X_train.columns
mutual_info.sort_values(ascending=False)
```

```python
mutual_info.sort_values(ascending=False).plot.bar(figsize=(15,5))
```

```python
from sklearn.feature_selection import SelectKBest
```

##### After the feature engineering, we'll use the best 5 features to predict the Life Expectancy

```python
sel_five_cols = SelectKBest(mutual_info_regression, k=5)
z= sel_five_cols.fit(X_train, y_train)
imp_cols = X_train.columns[z.get_support()]
```

```python
X_train=X_train[list(imp_cols)]
```

### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model=LinearRegression()
```

```python
lr_model.fit(X_train,y_train)
```

```python
X_test=X_test[list(imp_cols)]
```

```python
predictions=lr_model.predict(X_test)
```

```python
predictions
```

```python
print('This is regration accuracy score:')
lr_model.score(X_test, y_test)
```

```python
model1 = lr_model.score(X_test, y_test)
```

```python
best_model = []
best_model.append(model1)
```

```python
prediction_of_test = lr_model.predict(X_test)
plt.scatter(y_test, prediction_of_test, color='black')

x0 = min(y_test)
x1 = max(y_test)
y0 = min(prediction_of_test)
y1 = max(prediction_of_test)
plt.plot([x0,x1],[y0,y1],color="red")

plt.xlabel("Input Data")
plt.ylabel("Life Expectancy")
plt.title("Input Data Vs Life Expectancy")

print('Linear Regration of Input Data vs Life Expectancy:')
print('')
plt.show()
```

*Thus by looking at the graph, we can see that there are outliers in our data, if we want to make the predictions more precise we can elimate these outliers and then re-train the model to get more accurate predictions*

### Random Forest
